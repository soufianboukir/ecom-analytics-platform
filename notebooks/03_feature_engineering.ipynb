# Feature engineering
Feature engineering is the process of using domain knowledge to transform raw data into informative features that enhance machine learning model performance. It involves creating, selecting, and transforming input variables (features) to improve model accuracy, handle missing data, and reduce overfitting. Key techniques include encoding categorical data, normalizing numerical ranges, handling missing values, and generating new features based on business logic.

```python
# import necessary libs and data

import pandas as pd
df = pd.read_csv("../data/processed/amazon_sales.csv")
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])

df['Year'] = df['OrderDate'].dt.year
df['Month'] = df['OrderDate'].dt.month
df['Dayofweek'] = df['OrderDate'].dt.dayofweek
df['IsWeekEnd'] = df['Dayofweek'] >= 5

# shipping ratio
df['ShippingRatio'] = (df['ShippingCost'] / df['TotalAmount']).round(2)

# number of orders made by a customer
df['CustomerOrderCount'] = df.groupby('CustomerID')['OrderID'].transform('count')

# Customer average spend
df['CustomerAvgSpend'] = df.groupby('CustomerID')['TotalAmount'].transform('mean').round(2)

# Category average price
df['CategoryAvgPrice'] = df.groupby('Category')['UnitPrice'].transform('mean').round(2)

df.to_csv('../data/processed/amazon_sales_final.csv')
```
