# Data cleaning 

Data cleaning is the process of preparing data for analysis by removing or fixing data that is incorrect, incomplete, irrelevant, or duplicated within a dataset. It’s one of the important stages of machine learning. It plays a significant part in building a model.

## About the amazon sales dataset

This dataset contains 100,000 synthetic Amazon-style e-commerce sales transactions, designed to closely resemble real-world online retail behavior. With 20 clean and well-structured columns, it captures detailed information about customers, products, pricing, payments, logistics, and order outcomes.

```python
import pandas as pd


df = pd.read_csv("../data/raw/amazon_sales.csv")

print("Shape: ", df.shape)
print("Info about dataset: ", df.info())
print(df.describe())

df.head(10)
```

```python
# print sum of null values per column
print(df.isnull().sum())

# remove duplicates
df.drop_duplicates(inplace=True)

# remove impossible values
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]
df = df[(df['Discount'] >= 0) & (df['Discount'] <= 1)]

# save cleaned data as csv
df.to_csv('../data/processed/amazon_sales.csv', index=False)
```
